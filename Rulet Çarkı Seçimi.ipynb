```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
populasyon = ["A", "B", "C", "D", "E"]
uygunluk_degerleri = [10, 20, 30, 25, 15]
```

```python
# Rulet çarkı seçimi
def rulet_carki_secimi(populasyon, uygunluk_degerleri):
    toplam_uygunluk = sum(uygunluk_degerleri)
    olasiliklar = [deger / toplam_uygunluk for deger in uygunluk_degerleri]
    kümülatif_olasiliklar = np.cumsum(olasiliklar)
    rastgele_deger = np.random.rand()
    
    for i, kümülatif_olasilik in enumerate(kümülatif_olasiliklar):
        if rastgele_deger <= kümülatif_olasilik:
            return populasyon[i]
```

```python
# Seçim işlemi
secilen_bireyler = [rulet_carki_secimi(populasyon, uygunluk_degerleri) for _ in range(50)]
```

```python
# Seçim sonuçlarını sayma
secim_sayilari = {birey: secilen_bireyler.count(birey) for birey in populasyon}
```

```python
# Sonuçları pasta grafiği ile görselleştirme
plt.pie(secim_sayilari.values(), labels=secim_sayilari.keys(), autopct='%1.1f%%', startangle=90)
plt.title("Rulet Çarkı Seçimi Sonuçları (Pasta Grafiği)")
plt.axis("equal")  
plt.show()
```

```python
# Seçim sayılarını yazdırma
print("Seçim Sonuçları:")
for birey, sayi in secim_sayilari.items():
    print(f"{birey}: {sayi}")
```
